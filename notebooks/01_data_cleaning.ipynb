# 1. Очистка и переработка данных
**Цель:** Загрузить данные, удалить мусор и подготовить датасет для дальнейшего анализа.

**Источник данных:** Циан, апрель 2025 года, 1399 объявлений о продаже квартир в Москве. 

1.1 Импортирование библиотек

```python
import pandas as pd
import matplotlib.pyplot as plt
```

1.2 Загрузга данных

```python
df = pd.read_csv('../data/cian_flat_sale_1_57_moskva_19_Apr_2025_13_59_22_202458.csv', sep=';')
print("Количество строк и колонок")
print(df.shape)
```

1.3 Вывод первых 5 строк

```python
df.head(5)
```

1.4 Структура данных

```python
print("Типы данных в каждой колонке")
print(df.dtypes)
```

1.5 Вывод пропусков в каждой колонке

```python
print("Колличество пустых значений")
print(df.isnull().sum())
```

1.6 Удаление ненужных колонок и строк без цен

```python
df = df.drop(['url', 'phone', 'object_type', 'heating_type'], axis=1)
df = df.dropna(subset=['price'])
print(f"После удаления пустых цен: {len(df)} строк осталось")
```

1.7 Создание колонки "Цена за квадратный метр" и удаление выбросов

```python
df['price_per_m2'] = df['price'] / df['total_meters']
df = df[(df['price_per_m2']>200000)&(df['price_per_m2']<3000000)]
print(f"После удаления выбросов: {len(df)} строк осталось")
```

1.8 Чистка районов от мусора

```python
df['district_len'] = df['district'].str.len()
df = df[df['district_len'] < 20]
df = df.drop(['district_len'], axis=1)
print("Муссор в названиях районов удалён")
```

1.9 Медианная цена за квадратный метр

```python
median_price = df['price_per_m2'].median()
print("Медиана за квадратный метр:")
print(median_price)
```

1.10 Сохранение очищенного датасета

```python
df.to_csv('../data/cian_final_cleaned.csv', index=False)
print("Очищенные данные сохранены в data/cian_cleaned.csv")
```
